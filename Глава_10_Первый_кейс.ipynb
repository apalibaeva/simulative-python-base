**Примечание**

Не все файлы подходящие по названию, будут подходить по содержанию. Там может оказаться лишний столбец, например. Ориентируйтесь на столбцы из чека выше - это то, что вас интересует. Остальные столбцы можно просто отбросить.

**Важно**: разделителем файла на выходе должна быть запятая.

```python
import re, glob, os
def process_files(src_folder, dest_folder):
  if not os.path.exists(dest_folder):
      os.makedirs(dest_folder)
  new_path = os.path.join(dest_folder, 'combined_data.csv')

  files_list = glob.glob(src_folder + '/*.csv')

  required_columns = ['date', 'product', 'store', 'cost']
  header = [] #для проверки наличия необходимых столбцов в каждом файле
  combined_data =[]

  for el in files_list:
    if re.match(r'\d{4}-\d{2}-\d{2}-\d{2}-\d{2}-\d+.csv', el.split('/')[1]):
      with open(el, 'r', newline='', encoding='utf-8') as infile:
        reader = csv.DictReader(infile, delimiter=';')

        # Проверяем наличие необходимых столбцов
        if any(col in reader.fieldnames for col in required_columns):
            # Если заголовки еще не записаны, записываем их
            if not header:
                header = required_columns  # Запоминаем нужные заголовки

            # Записываем данные, выбирая только нужные столбцы
            for row in reader:
                combined_data.append({col: row[col] for col in required_columns if col in row})

  sort_columns = ['date', 'product']
  # Сортируем данные по указанным столбцам
  combined_data.sort(key=lambda x: (x[sort_columns[0]], x[sort_columns[1]]))

  # Сохранение объединенных данных в новый CSV файл
  with open(new_path, 'w', newline='', encoding='utf-8') as outfile:
      writer = csv.DictWriter(outfile, fieldnames=header)
      writer.writeheader()  # Записываем заголовки
      writer.writerows(combined_data)  # Записываем отсортированные данные


src_folder = 'reports-main'
dest_folder = 'comb_reports'
process_files(src_folder, dest_folder)
```
